## Plan

1. Dynamic programming for any n, m
2. For 114 -- F(3, 50), iterate over F(4, i) until reach 1000000

```python
F_cache = {}

def F(m, n) :
    if m > n :
        return 1
        
    if (m, n) in F_cache :
        return F_cache[(m, n)]
    
    count = 1
    for start in range(n - m + 1) :
        for block_size in range(m, n - start + 1) :
            # print (start, block_size), m, n, count, n - block_size - start - 1
            count += F(m, n - block_size - start - 1)
            
    F_cache[(m, n)] = count
    
    print (m, n), count
    return count

F(3, 7), F(3, 29), F(3, 30), F(10, 56), F(10, 57)
```
