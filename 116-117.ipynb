## Plan

1. Dynamic programming for any block length set
2. Calculate answers

```python
F_cache = {}

def F(n, block_set) :
    if (n, block_set) in F_cache :
        return F_cache[(n, block_set)]
    if n == 0 :
        return 1
    if n < min(block_set) :
        return 0
    
    count = 0
    for block_size in block_set :
        count += F(n - block_size, block_set)
            
    F_cache[(n, block_set)] = count
    
    print (n, block_set), count
    return count

F(5, (1, 2)) - 1, F(5, (1, 3)) - 1, F(5, (1, 4)) - 1, F(5, (1, 2, 3, 4))
```
