### Iterate over all sets and check if they are primes

```python
class memoize:
  def __init__(self, function):
    self.function = function
    self.memoized = {}

  def __call__(self, *args):
    try:
      return self.memoized[args]
    except KeyError:
      self.memoized[args] = self.function(*args)
      return self.memoized[args]
```

```python
import itertools

digits = (1, 2, 3, 4, 5, 6, 7, 8, 9)

def list_powerset(lst):
    # the power set of the empty set has one element, the empty set
    result = [[]]
    for x in lst:
        # for every additional element in our set
        # the power set consists of the subsets that don't
        # contain this element (just take the previous power set)
        # plus the subsets that do contain the element (use list
        # comprehension to add [x] onto everything in the
        # previous power set)
        result.extend([subset + [x] for subset in result])
    return result

@memoize
def primes_with_digits(digits) :
    count = 0
    for p in itertools.permutations(digits) :
        p = int(''.join(map(str, p)))
        if p in primes :
            # print 'Prime:',  p
            count += 1

    # print digits, count
    return count 

@memoize
def prime_set_count(digits) :
    if len(digits) == 0 :
        return 1
    count = 0
    for first_value_digits in list_powerset(digits) :
        if digits[0] not in first_value_digits :
            continue
        first_value_digits = tuple(sorted(first_value_digits))
        if sum(first_value_digits) % 3 == 0 and first_value_digits != (3,) :
            continue
    
        other_values_digits = tuple(sorted(set(digits) - set(first_value_digits)))
        primes_with_digits(first_value_digits)
        prime_set_count(other_values_digits)
        print "count: ", first_value_digits, other_values_digits, primes_with_digits(first_value_digits),  prime_set_count(other_values_digits)
        count += primes_with_digits(first_value_digits) * prime_set_count(other_values_digits)
    return count
```
