### Primes under 25000 :

```python
%time 

eratosphen = [True] * max_n
i = 2

primes = []

while i < max_n :
    if eratosphen[i] : 
        if i < 100 :
            print i
        
        primes.append(i)
        for not_prime in range(2 * i, max_n, i) :
            eratosphen[not_prime] = False
    i += 1


len(primes)
```

### Iterate over all producting-sum numbers from 4 and up

```python
primes_set = set(primes)
```

```python
def prime_divisors(n, primes) :
    divisors = []
    
    while n != 1 :
        for p in primes :
            count = 0
            while n % p == 0 :
                count += 1
                n /= p
            if count : 
                divisors.append((p, count))

            if n == 1 : 
                break
    return divisors

prime_divisors(24, primes)
```

```python
def make_value(prime_power_pairs) :
    value = 1
    for p, power in prime_power_pairs :
        value *= p ** power
        
    return value

def decrease_powers(divisors, max_divisors) :
    new_divisors = list(divisors)
    new_divisors[-1] = (divisors[-1][0], divisors[-1][1] - 1)
    for i in range(-1, -len(divisors) - 1, -1) :
        if new_divisors[i][1] < 0 :
            new_divisors[i] = max_divisors[i]
            new_divisors[i - 1] = (new_divisors[i - 1][0], new_divisors[i - 1][1] - 1)
    return new_divisors

def powers_more_then(a, b) :
    assert(len(a) == len(b))
    for x, y in zip(a, b) :
        if a < b : 
            return False
        if a > b :
            return True
    return False

def substract_divisors(a, b) :
    return [(x[0], x[1] - y[1]) for x, y in zip(a, b)] 

def divisors_sets_iterable(divisors, previous) :
    current_divisors = list(divisors)
        
    while True :
        if all([v[1] == 0 for v in current_divisors]) :
            break
            
        if powers_more_then(current_divisors, previous) :
            current_divisors = decrease_powers(current_divisors, divisors)
            continue
            
        remainder_divisors = substract_divisors(divisors, current_divisors)
        #print current_divisors, remainder_divisors
        
        for divisor_set in divisors_sets_iterable(remainder_divisors, current_divisors) :
            yield [make_value(current_divisors)] + divisor_set
        if all([v[1] == 0 for v in remainder_divisors]) :
            yield [make_value(current_divisors)]
        current_divisors = decrease_powers(current_divisors, divisors)


    

def divisors_sets(n, primes) :
    divisors = prime_divisors(n, primes)

    return divisors_sets_iterable(divisors, divisors)

for x in divisors_sets(125, primes) :
    print "set: ", x
```

```python
k_discovered = {}
count_k_discovered = 0
k_max = 12000
stop_iterating = False 

for n in range(4, max_n) :
    if n in primes_set :
        continue
    if stop_iterating : 
        break
    for divisors in divisors_sets(n, primes) :
        if len(divisors) == 1 : 
            continue

        x = n - sum(divisors)
        k = len(divisors) + x
        if k <= k_max and k not in k_discovered :
            k_discovered[k] = n
            print n, divisors, k
            count_k_discovered += 1
            if count_k_discovered >= k_max - 1: 
                stop_iterating = True
                break
                
print sum(set(k_discovered.values()))
```
